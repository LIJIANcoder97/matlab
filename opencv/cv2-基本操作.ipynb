### 读取视频
- cv2.VideoCapture 可以捕获摄像头，用数字控制不同设备。
- 视频文件直接指定路径。

```python
vc=cv2.VideoCapture('miao.webm')
```

```python
if vc.isOpened():
    open,frame=vc.read()# 读一帧，返回帧是否存在和帧内容
else:
    open=False
```

```python
while open:
    open ,frame = vc.read()
    if frame is None:
        break
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow('result',gray)
    cv2.waitKey(10)
vc.release()
cv2.destroyAllWindows()
```

### 边界填充

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(232),plt.imshow(img,'gray'),plt.title('原图')
plt.show()
```

```python
top_size,bottom_size,left_size,right_size = (50,50,50,50)
```

```python
replicata = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_REPLICATE)
plt.imshow(replicata,'gray'),plt.title('REPLICATE')
plt.show()
```

```python
reflect = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_REFLECT)
plt.imshow(reflect,'gray'),plt.title('REPLICATE')
plt.show()
```

```python
reflect_101 = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_REFLECT_101)
plt.imshow(reflect_101,'gray'),plt.title('REPLICATE')
plt.show()
```

```python
wrap = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_WRAP)
plt.imshow(wrap,'gray'),plt.title('REPLICATE')
plt.show()
```

```python
constant = cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_CONSTANT,value=0)
plt.imshow(constant,'gray'),plt.title('REPLICATE')
plt.show()
```

图像融合

```python
img = cv2.imread("cat.jpg")
r,g,b=cv2.split(img)
cat=cv2.merge((b,g,r))
img = cv2.imread("dog.jpg")
r,g,b=cv2.split(img)
dog=cv2.merge((b,g,r))
```

```python
plt.imshow(cat)
plt.show()
```

```python
plt.imshow(dog)
plt.show()
```

```python
new=cat+dog
```

```python
cat=cv2.resize(cat,(251,237))
```

```python
plt.imshow(cat)
plt.show()
```

```python
new=cat+dog
```

```python
plt.imshow(new)
plt.show()
```

```python
gray = cv2.cvtColor(new,cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show()
```

```python
new=cv2.addWeighted(cat,0.4,dog,0.6,0)# 模型融合
plt.imshow(new)
plt.show()
```
