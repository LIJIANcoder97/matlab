```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 梯度运算

### Sobel 算子

GX=右-左，GY=下-上

```python
img = cv2.imread("car.png")
r,g,b=cv2.split(img)
img=cv2.merge((b,g,r))
plt.imshow(img)
plt.show()
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
plt.imshow(sobelx)
plt.show()
```

sobel 算子检测边缘，图像先转换为灰度图

```python
img = cv2.imread("car.png")
r,g,b=cv2.split(img)
img=cv2.merge((b,g,r))
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray")
plt.show()
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
sobelx = cv2.convertScaleAbs(sobelx) #对输出进行绝对值处理，避免差值的负数被截断为0
plt.imshow(sobelx,cmap="gray")
plt.show()
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
sobely = cv2.convertScaleAbs(sobely) #对输出进行绝对值处理，避免差值的负数被截断为0
plt.imshow(sobely,cmap="gray")
plt.show()
```

```python
# GX与GY相加
sobelxy = cv2.addWeighted(sobelx,0.5,sobely,0.5,0)
plt.imshow(sobelxy,cmap="gray")
plt.show()
```

```python
### 对边界进行增强
ret,sobelxy = cv2.threshold(sobelxy,127,255,cv2.THRESH_BINARY)# 大于127为255，小于127为0
plt.imshow(sobelxy,cmap="gray")
plt.show()
```

### Scharr算子
- GXGY比Sobel算子的大

```python
img = cv2.imread("car.png")
r,g,b=cv2.split(img)
img=cv2.merge((b,g,r))
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray")
plt.show()
```

```python
scharrx = cv2.Scharr(img,cv2.CV_64F,1,0)
scharrx = cv2.convertScaleAbs(scharrx)
scharry = cv2.Scharr(img,cv2.CV_64F,0,1)
scharry = cv2.convertScaleAbs(scharry)
scharrxy = cv2.addWeighted(scharrx,0.5,scharry,0.5,0)
plt.imshow(scharrxy,cmap="gray")
plt.show()
```

```python
### 对边界进行增强
ret,scharrxy = cv2.threshold(scharrxy,127,255,cv2.THRESH_BINARY)# 大于127为255，小于127为0
plt.imshow(scharrxy,cmap="gray")
plt.show()
```

### laplacian算子 拉普拉斯算子
- 考虑2阶导，对变化比其他算子更敏感，但对噪音点更敏感

```python
img = cv2.imread("car.png")
r,g,b=cv2.split(img)
img=cv2.merge((b,g,r))
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray")
plt.show()
```

```python
### 先去噪，如果不先去噪，效果极差，噪声指车牌的反光
ret,img = cv2.threshold(img,170,255,cv2.THRESH_BINARY)# 大于127为255，小于127为0
plt.imshow(img,cmap="gray")
plt.show()
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian)
plt.imshow(laplacian,cmap="gray")
plt.show()
```

对其他算字进行去噪处理，效果会变差

## 边缘检测

### Canny 边缘检测
- 使用高斯滤波，平滑图像，消除噪声
- 计算像素的梯度强度和方向
- 应用非极大值抑制，消除边缘检测带来的杂散响应
- 应用双阈值检测确定真实和潜在的边缘
- 抑制孤立弱边缘完成边缘检测
